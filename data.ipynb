```python
# Dependencies
from GoogleNews import GoogleNews
from datetime import date, timedelta
import pandas as pd
```

```python
df = pd.read_csv('tickers.csv')

tickers = df['Symbol'].values.tolist()
# tickers = tickers[490:504]

print(tickers)
print(len(tickers))
```

```python
# Google News Parameters Setup
googlenews = GoogleNews()

titles = []
sort = False
for ticker in tickers:
    googlenews.search(ticker + ' Good Stock')
    news = googlenews.results(sort=sort)
    googlenews.clear()
    for articles in news[0:9]:
        titles.append([articles['title'], 1])
        # print(articles['title'])

    googlenews.search(ticker + ' Bad Stock')
    news = googlenews.results(sort=sort)
    googlenews.clear()
    for articles in news[0:9]:
        titles.append([articles['title'], -1])
        # print(articles['date'])
    print(ticker)
```
